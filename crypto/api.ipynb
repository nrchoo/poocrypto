```python
def query_mc_by_time(interval,limit):
    '''Function is used to query market cap by date on timely table. Returns dataframe'''
    conn = sqlite3.connect('/Users/nathanchoo/cryptodb/crypto.db')
    cur = conn.cursor()
    #Ideally these would aggregate all times of all coins by some factor. For now we can inject every 5 minutes to agg the hour
    query_unique_times = "SELECT DISTINCT(date) FROM CMC_FIVE order by date asc"
    cur.execute(query_unique_times)
    q2 = cur.fetchall()
    q2 = [i[0] for i in q2]
    q2 = q2[interval - 1::interval] #chops up the data into aggregate intervals.
    #conn.commit()
    #period = len(q)/interval
    
    ##########
    ed = {}
    date_list = q2
 
    for date in date_list:
        query_statement = '''SELECT * FROM CMC_FIVE WHERE date = ''' + "'" + str(date) + "'"  + ''' ORDER BY marketcap desc limit ''' + str(limit)
        cur.execute(query_statement)
        q = cur.fetchall()
        ed[q[0][1].split(' ')[1]] = [i[3] for i in q]
    
    return ed
```
